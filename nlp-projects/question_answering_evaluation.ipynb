Now that you have a decent context text with an accompanying test suite of questions and desired answers, write code that automatically evaluates the QA model on your data. Include the following metrics in your automatic quantitative evaluation:

- Exact match: the span proposed by the model matches exactly the desired span. - No match: the rest of the spans, where the proposed span and desired span do not overlap at all. Indicate the absolute number and relative proportion of such spans.


This function identifies the exact character spans of the gold answers within the context. When an anchor is provided, it ensures the correct occurrence is selected; for unanswerable questions, span values are set to None.

```python
# 1: code to obtain span gold answers

updated_pairs = []  # to store updated qa pairs with gold spans

for qa_pair in qa_pairs:
    gold_answer = qa_pair["gold"]

    # Unanswerable question
    if gold_answer == "":
        qa_pair["gold_start"] = None
        qa_pair["gold_end"] = None
        updated_pairs.append(qa_pair)
        continue

    # If anchor exists → use it to locate correct occurrence
    if "anchor" in qa_pair:
        anchor = qa_pair["anchor"]
        anchor_start = context.find(anchor)

        if anchor_start == -1:
            raise ValueError(f"Anchor not found in context: {anchor}")

        relative_index = anchor.find(gold_answer)
        if relative_index == -1:
            raise ValueError(f"Gold answer not inside anchor: {gold_answer}")

        start_index = anchor_start + relative_index  # absolute start index of gold in context

    else:
        # fallback: find gold directly
        start_index = context.find(gold_answer)
        if start_index == -1:
            raise ValueError(f"Gold answer '{gold_answer}' not found in context.")

    end_index = start_index + len(gold_answer)  # end index of gold in context

    # append span info
    qa_pair["gold_start"] = start_index
    qa_pair["gold_end"] = end_index

    updated_pairs.append(qa_pair)
```

```python
# 2: function to classify match - exact, narrow, broad, partial, no_match

for qa_pair in updated_pairs:
    question = qa_pair["question"]
    gold_start = qa_pair["gold_start"]
    gold_end = qa_pair["gold_end"]
    # Get model prediction
    result = qa_model(question=question, context=context)
    pred_start = result["start"]
    pred_end = result["end"]
    # Unanswerable gold
    if gold_start is None:
        # if model still returns a span, it's a false answer
        match_type = "no_match"
    # Exact match
    elif pred_start == gold_start and pred_end == gold_end:
        match_type = "exact"
    # Narrow match (prediction inside gold)
    elif pred_start >= gold_start and pred_end <= gold_end:
        match_type = "narrow"
    # Broad match (prediction covers gold)
    elif pred_start <= gold_start and pred_end >= gold_end:
        match_type = "broad"
    # Partial overlap
    elif pred_start < gold_end and pred_end > gold_start:
        match_type = "partial"
    # No overlap
    else:
        match_type = "no_match"

    qa_pair["match_type"] = match_type
    qa_pair["predicted_answer"] = result["answer"]
    qa_pair["score"] = result["score"]

    print(f"\nQ{qa_pair['id']}: {question}")
    print(f"Gold span : {context[gold_start:gold_end] if gold_start is not None else 'N/A'}")
    print(f"Predicted : {result['answer']}")
    print(f"Match type: {match_type}")
    print(f"Score     : {result['score']:.3f}")
```

```python
exact = 0
narrow = 0
broad = 0
partial = 0
no_match = 0

for q in updated_pairs:
    if q["match_type"] == "exact":
        exact += 1
    elif q["match_type"] == "narrow":
        narrow += 1
    elif q["match_type"] == "broad":
        broad += 1
    elif q["match_type"] == "partial":
        partial += 1
    else:
        no_match += 1

total = len(updated_pairs)

print("\nRESULTS")
print(" ")
print("Exact:", exact, f"({exact/total*100:.1f}%)")
print("Narrow:", narrow, f"({narrow/total*100:.1f}%)")
print("Broad:", broad, f"({broad/total*100:.1f}%)")
print("Partial:", partial, f"({partial/total*100:.1f}%)")
print("No match:", no_match, f"({no_match/total*100:.1f}%)")
```

### Task 9
Study the score value that is returned together with answer spans. How well does this value reflect the confidence or reliability of the correctness of the answer? Comment!

The score returned by the QA model represents how confident the model is that the extracted span answers the question. In many cases, higher scores correspond to correct answers. For example, factual questions with explicit answers in the text (e.g. Q2, Q6, Q9) receive high scores and correct predictions, suggesting that the score is a reasonable indicator of confidence when the answer is clearly expressed.

However, the score is not a reliable indicator of correctness in all cases. Some incorrect answers receive relatively high scores. For instance, in Q5 the model predicts “Guillermo” with a very high score even though the gold answer is “Guillermo Garcia” (Moreover, the extracted name comes from a different span than the one marked in the gold label; therefore, the prediction is counted as a mismatch even though it represents a narrowed version of the correct answer.). Similarly, in Q10 the model confidently returns “Oscar” instead of “Jude and Oscar”. This shows that the model may be highly confident in a partially correct or incomplete span.

Conversely, some correct answers receive low scores (e.g., Q11), indicating that a low score does not necessarily imply an incorrect prediction.

Scores become particularly unreliable for nuanced or inference-based questions. In Q17 and Q18 the model assigns low scores to incorrect spans, but the low confidence does not help identify the correct answer because the model fails to retrieve it. For unanswerable questions (Q19–Q20), the model still returns spans with low scores, suggesting uncertainty but not explicitly signaling that no answer exists.

Overall, the score is a useful indicator of confidence when answers are explicit and directly retrievable. However, it should not be interpreted as a direct measure of correctness.

```python
def classify_match(model, match_key_prefix):
    for qa_pair in updated_pairs:
        question = qa_pair["question"]
        gold_start = qa_pair["gold_start"]
        gold_end = qa_pair["gold_end"]

        result = model(question=question, context=context)

        pred_start = result["start"]
        pred_end = result["end"]
        pred_answer = result["answer"]

        if gold_start is None:
            if pred_answer == "":
                match_type = "impossible_correct" 
            else:
                match_type = "no_match"
        else:
            if pred_start == gold_start and pred_end == gold_end:
                match_type = "exact"
            elif pred_start >= gold_start and pred_end <= gold_end:
                match_type = "narrow"
            elif pred_start <= gold_start and pred_end >= gold_end:
                match_type = "broad"
            elif pred_start < gold_end and pred_end > gold_start:
                match_type = "partial"
            else:
                match_type = "no_match"

        qa_pair[f"match_type_{match_key_prefix}"] = match_type
        qa_pair[f"predicted_answer_{match_key_prefix}"] = pred_answer
        qa_pair[f"score_{match_key_prefix}"] = result["score"]

    exact = narrow = broad = partial = no_match = impossible_correct = 0

    for q in updated_pairs:
        m = q[f"match_type_{match_key_prefix}"]

        if m == "exact":
            exact += 1
        elif m == "narrow":
            narrow += 1
        elif m == "broad":
            broad += 1
        elif m == "partial":
            partial += 1
        elif m == "impossible_correct":
            impossible_correct += 1
        else:
            no_match += 1

        total = len(updated_pairs)

    print("\nRESULTS")
    print(" ")
    print("Exact:", exact, f"({exact/total*100:.1f}%)")
    print("Narrow:", narrow, f"({narrow/total*100:.1f}%)")
    print("Broad:", broad, f"({broad/total*100:.1f}%)")
    print("Partial:", partial, f"({partial/total*100:.1f}%)")
    print("No match:", no_match, f"({no_match/total*100:.1f}%)")
    print("Impossible (correct):", impossible_correct, f"({impossible_correct/total*100:.1f}%)")
```

```python
classify_match(qa_model, "distilbert")
classify_match(qa_model_2, "roberta")
classify_match(qa_model_3, "conscious")
```
