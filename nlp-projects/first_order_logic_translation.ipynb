### Task 1: Translate the following sentences into **propositional logic**

```python
read_expr = nltk.sem.Expression.fromstring

# sentences to be translated into logic
'''
- Harald sings and dances.
- Harald doesn't walk or run.
- I will win if you don't win.
- If you lose, it is not the case that I care.
- If you listen when I talk, then I will listen if you talk.
'''
read_expr('HS & HD') # two propositional symbols for "Harald sings" and "Harald dances" linked by conjunction (boolean AND)
read_expr('-(HW | HR)') # two propositional symbols for "Harald walks" and "Harald runs" linked by disjunction (boolean OR) and negated to express "doesn't walk or run"
read_expr('-Yw -> Iw') # propositional symbols for "you win" and "I win" linked by implication (->) and negated to express "if you don't win, then I will win"
read_expr('L -> -C') # propositional symbols for "you lose" and "I care" linked by implication (->) and negated to express "if you lose, it is not the case that I care"
read_expr('LwT -> (T -> Il)') # propositional symbols for "you listen when I talk", "I will listen if you talk" linked by implication (->) to express "if you listen when I talk, then I will listen if you talk"

'''I had doubts about combining “If you listen when I talk” into a single propositional symbol.'''
```

### Task 2: Translate the following sentences into *predicate-argument formula* of **first order logic**

```python
# sentences to be translated into first-order logic
'''
- Harald sings and dances.
- Harald likes Jim and Kim.
- Jim likes himself and so does Kim.
- Harald and Kim hug each other.
- Harald is a friendly person, but Jim is not.
'''

read_expr('sing(harald) & dance(harald)') # two atomic formulas for "Harald sings" and "Harald dances" linked by conjunction 
read_expr('like(harald, jim) & like(harald, kim)') # alternative: all x.((x = jim | x = kim) -> Like(harald,x)) too complicated and less intuitive than the first one
# The sentence "Harald likes Jim and Kim." can be translated using two atomic formulas linked by conjunction, which is more straightforward and easier to understand than using a universal quantifier and implication. The first translation directly states that Harald likes both Jim and Kim, while the second translation introduces unnecessary complexity by using a universal quantifier and implication to express the same idea.

read_expr('like(jim, jim) & like(kim, kim)') # alternative: all x.((x = jim | x = kim) -> Like(x,x)) 
# The sentence “Jim likes himself and so does Kim.” can be represented with two simple atomic statements joined by a conjunction. This approach is clearer and more direct than using a universal quantifier with an implication. The straightforward version explicitly states that Jim likes himself and that Kim likes herself, whereas the quantified version expresses the same meaning in a more complicated way without adding useful information.

read_expr('hug(harald, kim) & hug(kim, harald)') # it is not so different from the previous sentence.
read_expr('friendly(harald) & -friendly(jim)')  # by looking at the sentence, it would be clearer if we specified that both Harald and Jim are people, e.g. Person(harald) & Person(jim) & Friendly(harald) & -Friendly(jim), but this is not strictly necessary for the translation of the sentence itself.
```

### Task 3: Translate the following sentences into *quantified formulas* of **first order logic**

```python
# translating sentences with quantifiers
'''
- Everybody loves a puppy.
- Whales are mammals, but sharks are not.
- There is a person that Harald loves but who does not love him back.
- Somebody sings and dances.
- Nobody sings or dances.
- Jim is in love with exactly one person.
- Nobody else than Harald plays the piano.
'''

read_expr('all x. (person(x) -> exists y. (puppy(y) & loves(x,y)))') # the sentence "Everybody loves a puppy." can be translated as for all x, if x is a person, then there exists a y such that y is a puppy and x loves y"
read_expr('all x.(whale(x) -> mammal(x)) & all y.(shark(y) -> -mammal(y))') # the sentence "Whales are mammals, but sharks are not." can be translated as "for all x, if x is a whale, then x is a mammal, and for all y, if y is a shark, then y is not a mammal".
read_expr('exists x.(person(x) & loves(harald,x) & -loves(x,harald))') # the sentence "There is a person that Harald loves but who does not love him back." can be translated as "there exists an x such that x is a person, Harald loves x, and x does not love Harald". 
read_expr('exists x. (person(x) & sings(x) & dances(x))') # the sentence "Somebody sings and dances." can be translated as "there exists an x such that x is a person, x sings, and x dances" 
read_expr('-exists x.(sings(x) | dances(x))') # the sentence "Nobody sings or dances." can be translated as "there does not exist an x such that x sings or x dances". 
read_expr('exists x. (person(x) & loves(jim,x) & all y. ((person(y) & loves(jim,y)) -> (y = x)))') # "there exists an x such that x is a person, Jim loves x, and for all y, if y is a person and Jim loves y, then y is identical to x"
read_expr('all x.(piano(x) -> plays(harald,x)) & -exists y.(person(y) & y != harald & all z.(piano(z) -> plays(y,z)))') # the sentence "Nobody else than Harald plays the piano." can be translated as "for all x, if x is a piano, then Harald plays x, and there does not exist a y such that y is a person, y is not Harald, and for all z, if z is a piano, then y plays z". 
```

### Task 4: Translate the following verb phrases using **λ abstracts**

```python
# translating verb phrases with using lambda abstraction
'''
- drink tea and eat biscuits
- buy beer to everyone
- be given beer by Jim
- be remembered or forgotten by everyone
- be remembered by everyone and forgotten by no-one
'''
read_expr(r'\x.(drink(x, tea) & eat(x, biscuits))') # the verb phrase "drink tea and eat biscuits" can be translated as a lambda expression that takes an argument x and states that x drinks tea and eats biscuits.
read_expr(r'\x.(all y.(person(y) -> buy(x, beer, y)))') # the verb phrase "buy beer to everyone" can be translated as a lambda expression that takes an argument x and states that for all y, if y is a person, then x buys beer to y.
read_expr(r'\x.(give(jim, beer, x))') # the verb phrase "be given beer by Jim" can be translated as a lambda expression that takes an argument x and states that Jim gives beer to x.
read_expr(r'\x.(all y.(person(y) -> (remember(y,x) | forget(y,x))))') # the verb phrase "be remembered or forgotten by everyone" can be translated as a lambda expression that takes an argument x and states that for all y, if y is a person, then y either remembers x or forgets x.
read_expr(r'\x.(all y.(person(y) -> (remember(y,x) & -forget(y,x))))') # the verb phrase "be remembered by everyone and forgotten by no-one" can be translated as a lambda expression that takes an argument x and states that for all y, if y is a person, then y remembers x and does not forget x.
```

### Task 5: Lambda function application

```python
# In each of the following statements, a declaration of e1 is missing. Th task is to provide it

read_expr = nltk.sem.Expression.fromstring

'''dance(harald)'''

e1 = read_expr(r'\x.(dance(x))')
e2 = read_expr('harald')
e3 = nltk.sem.ApplicationExpression(e1, e2)
print(e3.simplify())

'''exists y.eat(jim, y)'''
e1 = read_expr(r'\x.(exists y.(eat(x, y)))') # the verb phrase "eat" can be translated as a lambda expression that takes an argument x and states that there exists a y such that x eats y.
e2 = read_expr('jim')
e3 = nltk.sem.ApplicationExpression(e1, e2)
print(e3.simplify())

'''fly(woody)'''
e1 = read_expr(r'\P.(P(woody))') # the verb phrase "fly" can be translated as a lambda expression that takes a predicate P and states that P applies to woody.
e2 = read_expr('fly')
e3 = nltk.sem.ApplicationExpression(e1, e2)
print(e3.simplify())

'''all x.(bird(x) -> fly(x))'''
e1 = read_expr(r'\Q.Q(\x.fly(x))') 
e2 = read_expr(r'\P. all x. (bird(x) -> P(x))')
e3 = nltk.sem.ApplicationExpression(e1, e2)
print(e3.simplify())

# e1 takes a predicate-taking function Q and applies it to the predicate fly.
# The lambda \x.fly(x) represents the property of flying, which Q expects as input.
# When Q is applied to this predicate, we obtain a sentence stating that all birds fly.
# No implication (->) is needed because we are performing functional application,
# not expressing a logical condition.
```
